```python
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn_extra.cluster import KMedoids
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

# Fit and plot KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
df['kmeans_cluster'] = kmeans.fit_predict(df_scaled)

# Fit and plot Hierarchical Clustering
agg_clustering = AgglomerativeClustering(n_clusters=3)
df['agg_cluster'] = agg_clustering.fit_predict(df_scaled)

# Fit and plot K-Medoids
kmedoids = KMedoids(n_clusters=3, random_state=42)
df['kmedoids_cluster'] = kmedoids.fit_predict(df_scaled)

# Plot clusters
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
sns.scatterplot(x='longitude', y='latitude', hue='kmeans_cluster', data=df, palette='viridis')
plt.title('KMeans Clustering')

plt.subplot(1, 3, 2)
sns.scatterplot(x='longitude', y='latitude', hue='agg_cluster', data=df, palette='viridis')
plt.title('Hierarchical Clustering')

plt.subplot(1, 3, 3)
sns.scatterplot(x='longitude', y='latitude', hue='kmedoids_cluster', data=df, palette='viridis')
plt.title('K-Medoids Clustering')

plt.show()

# Silhouette Score
silhouette_kmeans = silhouette_score(df_scaled, df['kmeans_cluster'])
silhouette_agg = silhouette_score(df_scaled, df['agg_cluster'])
silhouette_kmedoids = silhouette_score(df_scaled, df['kmedoids_cluster'])

# Summary
print(f"Silhouette Score (KMeans): {silhouette_kmeans}")
print(f"Silhouette Score (Hierarchical Clustering): {silhouette_agg}")
print(f"Silhouette Score (K-Medoids): {silhouette_kmedoids}")
```

```python
def plot_cluster_distribution(df, cluster_column, variable_column, title):
    plt.figure(figsize=(10, 6))
    sns.countplot(x=variable_column, hue=cluster_column, data=df, palette='viridis')
    plt.title(title)
    plt.xlabel(variable_column)
    plt.ylabel('Count')
    plt.show()

# Plot cluster distribution for race, armed, and threat level
plot_cluster_distribution(df, 'kmeans_cluster', 'race', 'Cluster Distribution by Race (KMeans)')
plot_cluster_distribution(df, 'agg_cluster', 'race', 'Cluster Distribution by Race (Hierarchical)')
plot_cluster_distribution(df, 'kmedoids_cluster', 'race', 'Cluster Distribution by Race (K-Medoids)')
```

```python
def plot_cluster_scatter(df, cluster_column, title):
    plt.figure(figsize=(15, 15))
    sns.scatterplot(x='longitude', y='latitude', hue=cluster_column, data=df, palette='viridis')

    # Plot circles around each cluster
    for cluster in df[cluster_column].unique():
        cluster_points = df[df[cluster_column] == cluster][['longitude', 'latitude']]
        center = cluster_points.mean()
        cov_matrix = cluster_points.cov()
        lambda_, v = np.linalg.eig(cov_matrix)
        angle = np.degrees(np.arctan2(*v[:, 0][::-1]))
        width, height = 2 * np.sqrt(lambda_)
        ellipse = plt.matplotlib.patches.Ellipse(xy=center, width=width, height=height, angle=angle, color='none', edgecolor='red', linewidth=2)
        plt.gca().add_patch(ellipse)

    plt.title(title)
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.legend()
    plt.show()

# Plot scatter plot with cluster circles for KMeans
plot_cluster_scatter(df, 'kmeans_cluster', 'KMeans Clustering with Cluster Circles')

# Silhouette Score
silhouette_kmeans = silhouette_score(df_scaled, df['kmeans_cluster'])
print(f"Silhouette Score (KMeans): {silhouette_kmeans}")
```
